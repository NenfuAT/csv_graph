```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 28

fname = "cut_nasu"
acc = pd.read_csv(fname+".csv",encoding = 'utf-8')

fig = plt.figure(figsize=(20, 20))

ax1 = fig.add_subplot(2, 1, 1)
ax1.grid(color='k', linestyle='dotted', linewidth=1, alpha=0.5, zorder=2)
ax1.set_title("生")
ax1.set_xlabel('time[s]')
ax1.set_ylabel('accelaration[$m/s^2$]')
x = ax1.plot((acc["time"]-acc["time"][0])/1000, acc["x"], label="x")
y = ax1.plot((acc["time"]-acc["time"][0])/1000, acc["y"], label="y")
z = ax1.plot((acc["time"]-acc["time"][0])/1000, acc["z"], label="z")
nolm=ax1.plot((acc["time"]-acc["time"][0])/1000, (acc["x"]**2 + acc["y"]**2 + acc["z"]**2)**0.5, label="nolm")
ax1.legend(handles = x + y +z+nolm)

ax2 = fig.add_subplot(2, 1, 2)
ax2.grid(color='k', linestyle='dotted', linewidth=1, alpha=0.5, zorder=2)
ax2.set_title("フィルタリング")
ax2.set_xlabel('time[s]')
ax2.set_ylabel('accelaration[$m/s^2$]')
x = ax2.plot((acc["time"]-acc["time"][0])/1000, acc["x"].rolling(window=3).mean(), label="x")
y = ax2.plot((acc["time"]-acc["time"][0])/1000, acc["y"].rolling(window=3).mean(), label="y")
z = ax2.plot((acc["time"]-acc["time"][0])/1000, acc["z"].rolling(window=3).mean(), label="z")
nolm=ax2.plot((acc["time"]-acc["time"][0])/1000, ((acc["x"]**2 + acc["y"]**2 + acc["z"]**2)**0.5).rolling(window=3).mean(), label="nolm")
ax2.legend(handles = x + y +z+nolm)
```
